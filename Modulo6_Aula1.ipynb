```python
categoricas = {'Genero','Categoria_de_renda','Grau_Escolaridade','Estado_Civil','Moradia','Ocupacao','Tamanho_Familia'}
binarias = {cat for cat in df_clientes_cadastrados_ID_unicos.columns if cat.rfind('Tem')!=-1}
discretas = {'Qtd_Filhos','Tamanho_Familia'}
continuas = set(df_clientes_cadastrados_ID_unicos.columns) - categoricas - binarias - discretas - {'ID_Cliente'}
print(f'categoricas: {categoricas}\nbinarias: {binarias}\ndiscretas: {discretas}\ncontinuas: {continuas}')
```
