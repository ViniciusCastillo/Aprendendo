```python
from matplotlib.markers import MarkerStyle
from IPython.core.display import Markdown
def modelar (df, df_teste, y_label, growth='linear', changepoints=None, n_changepoints=25, changepoint_range=0.8, yearly_seasonality='auto', 
             weekly_seasonality='auto', daily_seasonality='auto', holidays=None, seasonality_mode='additive', seasonality_prior_scale=10.0, 
             holidays_prior_scale=10.0, changepoint_prior_scale=0.05, mcmc_samples=0, interval_width=0.8, uncertainty_samples=1000, 
             stan_backend=None):
  modelo = Prophet(growth, changepoints, n_changepoints, changepoint_range, yearly_seasonality, weekly_seasonality, daily_seasonality, 
                   holidays, seasonality_mode, seasonality_prior_scale, holidays_prior_scale, changepoint_prior_scale, mcmc_samples, 
                   interval_width, uncertainty_samples, stan_backend)
  modelo.fit(df)
  df_futuro = modelo.make_future_dataframe(periods=len(df_teste))
  if growth=='logistic': df_futuro['cap'] = df['cap'].iloc[0];
  previsao = modelo.predict(df_futuro)
  fig = modelo.plot(previsao, xlabel='Data', ylabel='Casos novos', figsize=(15,8))
  a = add_changepoints_to_plot(fig.gca(), modelo, previsao)
  plt.plot(df_teste.ds, df_teste.y, '-', marker='.', color='orange')
  plt.plot(df.ds, df.y, '-', color="grey")
  plt.xticks(color='black')
  plt.yticks(color='black')
  return previsao
```

```python
def cria_DFs (dados, campo_y, inicio_teste, inicio_treino=0, campo_data='datahora'):
  df_teste = pd.DataFrame()
  df_teste['ds'] = dados[campo_data][inicio_teste:]
  df_teste['y'] = dados[campo_y][inicio_teste:]
  df = pd.DataFrame()
  df['ds'] = dados[campo_data][inicio_treino:inicio_teste]
  df['y'] = dados[campo_y][inicio_treino:inicio_teste]
  return df, df_teste
```

```python
df, df_teste = cria_DFs(dados_sp, 'casos', 400)
df['cap'] = 40000000 * 0.02
previsao = modelar(df, df_teste, 'Casos Acumulados', 'logistic', changepoint_range=0.9)
```

```python
df, df_teste = cria_DFs(dados_sp, 'casos_novos', 410, 290)
previsao = modelar(df, df_teste, 'Casos', changepoints=['2021-03-03'], changepoint_range=0.9)
```

```python
#Usando feriados especificos do municipio de SP

feriados_sp_2020 = [
'2020-02-25',
'2020-02-26',
'2020-04-10',
'2020-02-25',
'2020-02-26',
'2020-04-10',
'2020-04-21',
'2020-05-01',
'2020-05-20',
'2020-05-21', 
'2020-05-22',
'2020-05-25',
'2020-06-11',
'2020-09-07',
'2020-10-12',
'2020-11-02',
'2020-11-15',
'2020-12-25'
]

feriados_sp_2021 = [
'2021-01-01',
'2021-01-25',
'2021-02-15',
'2021-02-16',
'2021-02-17',
'2021-03-26',
'2021-03-29',
'2021-03-30',
'2021-03-31',
'2021-04-01',
'2021-04-02',
'2021-04-21',
'2021-05-01'
]


feriados_sp_2020 = pd.DataFrame({
  'holiday': 'sp_2020',
  'ds': pd.to_datetime(feriados_sp_2020)
})

feriados_sp_2021 = pd.DataFrame({
  'holiday': 'sp_2021',
  'ds': pd.to_datetime(feriados_sp_2021)
})

feriados = pd.concat((feriados_sp_2020, feriados_sp_2021))
feriados.head()
```

```python
df, df_teste = cria_DFs(dados_sp, 'casos_novos', 400)
previsao_add = modelar(df, df_teste, 'Casos', changepoint_prior_scale=10.0, holidays=feriados)
```

```python
df, df_teste = cria_DFs(dados_sp, 'casos_novos', 400)
previsao_mult = modelar(df, df_teste, 'Casos', changepoint_prior_scale=10.0, holidays=feriados, seasonality_mode='multiplicative')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
#Sazonalidade aditiva
print ("Sazonalidade por Adição\n")
print ("Mean Absolute Error: %.2f" % mean_absolute_error(pd.concat((df['y'],df_teste['y'])), previsao_add['yhat']))
print ("Mean Squared Error: %.2f" % mean_squared_error(pd.concat((df['y'],df_teste['y'])), previsao_add['yhat']))
print ("Root Mean Squared Error: %.2f" % mean_squared_error(pd.concat((df['y'],df_teste['y'])), previsao_add['yhat'], squared=False))
print ("Mean Absolute Percentage Error: %r" % mean_absolute_percentage_error(pd.concat((df['y'],df_teste['y'])), previsao_add['yhat']))
#Sazonalidade multiplicativa
print ("\n\nSazonalidade por Multiplicação\n")
print ("Mean Absolute Error: %.2f" % mean_absolute_error(pd.concat((df['y'],df_teste['y'])), previsao_mult['yhat']))
print ("Mean Squared Error: %.2f" % mean_squared_error(pd.concat((df['y'],df_teste['y'])), previsao_mult['yhat']))
print ("Root Mean Squared Error: %.2f" % mean_squared_error(pd.concat((df['y'],df_teste['y'])), previsao_mult['yhat'], squared=False))
print ("Mean Absolute Percentage Error: %r" % mean_absolute_percentage_error(pd.concat((df['y'],df_teste['y'])), previsao_mult['yhat']))
```
